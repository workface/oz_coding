__[문제1]__

__[문제2]__

10명이 각자 4개의 공을 가지고 있습니다.

운이 좋은 3명이 2개의 공을 더 받았습니다.

10명이 가진 모든 공의 총 개수를 구하세요.


```python
people = 10 #사람 10명
ball = 4 #각자 공 4개
lucky_people = 3 #운좋은 사람2명
lucky_people_per_ball = 2 #운좋은 사람이 얻은공 2개
print((people * ball) + (lucky_people * lucky_people_per_ball))
```

__[문제3]__

이상한 강박관념이 있는 나는 모든 시간을 분으로 계산하고 싶습니다.

대신 계산해주세요 !


```python
student = 5 #학생수
time = 3 #공부한 시간
print(student * time * 60)#1시간에 60분
```

__[문제5]__

어떤 정사각형의 넓이가 144cm²입니다.

정사각형의 한 변의 길이를 구하세요.


```python
area=144
#정사각형의 넓이를 구하는 방법은 한변*한변이므로 144의 제곱근을 구해 준다
print(area**0.5)
```
