__[문제3]__

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
s = "77three4one"

number_map = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9"
}


#결과를 저장할 변수
#영단어를 저장할 변수
# s값을 하나씩 순회한다.
# isdigit= 문자열이 숫자로만 되어있는지 확인하는 메서드 를 사용해 s 안의 값이 숫자인지 판별
# 숫자라면 result에 포함
# 아니라면 temp 라는 변수에 담는다
# temp안에 영어가 number_map에 있다면
# number_map[temp]를 해서 영어에 포함된 숫자를 result에 포함해준다 
#temp는 초기화 한다
#reselt 출력

result = "" #결과값을 저장할 변수

temp = "" #영단어를 저장할 변수

for i in s : #s값을 순회한다
    if i.isdigit() :
        result += i
    else :
        temp += i
        if temp in number_map :
            result += number_map[temp]
            temp = ""
print(result)
```

__[문제4]__

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다. 

 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = []

for i in range(len(arr1)) :
    row = []
    for j in range(len(arr1[i])) :
        row.append(arr1[i][j] + arr2[i][j])
    result.append(row)
print(result)
```

__[문제5]__

양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 합니다.   
예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.   
주어진 x를 이용해 x가 하샤드 수인지 아닌지 검사하는 코드를 작성해주세요

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
i = int(input('원하는 수 입력'))

digit_nums = 0 #자릿수 합 초기화
for digit in str(i) : # i 를 문자열로 바꿔 각자리를 순회
    digit_nums += int(digit) # digit_nums에 i의 각자리수 합을 할당
if i % digit_nums == 0 : # i를 digit_nums로 % 를 이용해 계산해 나머지가 없으면 하샤드 수가 맞고 아니면 하샤드수가 아니다.
    print(f'{i}은(는) 하샤드 수 입니다.')
else :
    print(f'{i}은(는) 하샤드 수가 아닙니다')
```

__[문제7]__

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
numbers = [5, 8, 1, 0, 6, 9] 
missing_num = []
for i in range(0, 10) :
    if i not in numbers :
        missing_num.append(i)
print(missing_num)
print(sum(missing_num))
```
