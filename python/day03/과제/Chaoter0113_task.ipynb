__[문제 1: 잔액 확인]__

현재 잔액은 1000원입니다. 현재 잔액을 출력하세요. 

(변수명 예시 : 잔액 - balance)

```python
balance = 1000
print(f'현재잔액은 {balance}원 입니다')
```

__[문제 2: 입금]__

가지고 있는 돈을 입금합니다.

입금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 입금 거래를 영수증 리스트에 기록하세요.

- 입금 금액은 input()을 이용해 받습니다.
- 영수증은 list()로 선언합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 영수증 리스트 - receipts, 입금 금액 - deposit_amount)


```python
deposit_amount = int(input("입금하실 금액을 입력해 주세요"))
print(f'입금하신 금액은 {deposit_amount}원 입니다')
now_balance = deposit_amount + balance #입금 후 잔액
print(f'현재 잔액은 {now_balance}원 입니다')
receipts = ("입금", deposit_amount, now_balance)
print(f'----영수증 ---- \n 내용 : {receipts[0]}\n 입금 금액 : {receipts[1]} \n 현재 잔액 : {receipts[2]}')
```

__[문제 3: 출금]__

사용자가 돈을 출금합니다. 출금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 출금 거래를 영수증 리스트에 기록하세요. 

- 출금 금액은 input()을 이용해 받습니다.
- 현재 잔액보다 출금 요청 금액이 많을 경우 현재 잔액 만큼만 출금되도록 합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 출금금액 - withdraw_amount)

```python
withdraw_amount = int(input(f'출금하실 금액을 입력해 주세요.'))
if withdraw_amount > now_balance :
    print(f'출금금액이 현재 잔액 보다 많습니다. {now_balance}원 만큼만 출금됩니다')
    withdraw_amount = now_balance
now_balance -= withdraw_amount
print(f'{withdraw_amount}원이 춤금되었습니다. 현재 잔액은{now_balance}입니다')
```
